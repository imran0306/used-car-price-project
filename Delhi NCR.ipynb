```python
# scrape all Cars urls
Car_urls_amd = []
url_amd=driver.find_elements(By.XPATH,"//*/div/div/div[2]/div[1]/div/h3/a")
for car in url_amd:
        Car_urls_amd.append(car.get_attribute("href"))
```

```python
Links_amd=Car_urls_amd
len(Car_urls_amd)
```

```python
# Make Empty List of the required data
Brand= []
Model= []
Make_Year= []
Fuel=[]
KMs_Driven=[]
Engine_Displacement=[]
No_Of_Owner=[]
Transmission=[]
Mileage=[]
Max_Power=[]
Torque=[]

Seats=[]
Color=[]
Gear_Box=[]
Drive_Type=[]
Steering_Type=[]
Front_Brake_Type=[]
Rear_Brake_Type=[]

Acceleration=[]
Engine_Type=[] 
Wheel_Base = []
Price= []
```

```python
for link in Car_urls_amd:
    driver.get(link)
  
    # Extracting Price 
    try:
        price = driver.find_element(By.XPATH,'//*[@id="vdp-main"]/div/div/div[1]/div[2]/div[1]/div/div[3]/span')      
        Price.append(price.text)
   
    except NoSuchElementException:
        Price.append('None')
        
    # Extracting Brand from xpath
    try:
        brand = driver.find_element(By.XPATH,'//*[@id="vdp-main"]/div/div/div[1]/div[2]/div[1]/div/div[1]')      
        Brand.append(brand.text)
    except NoSuchElementException:
        Price.append('None')  
        
    # Extracting make year
    try:
        year = driver.find_element(By.XPATH,'//*[@id="vdpCard-carOverview"]/ul/li[10]/div/div/span')      
        Make_Year.append(year.text)
    except NoSuchElementException:
        Make_Year.append('-')
        
        
# Extracting Fuel Type 
    try:
        fuel = driver.find_element(By.XPATH,'//*[@id="vdpCard-carOverview"]/ul/li[3]/div/div/span')      
        Fuel.append(fuel.text)
    except NoSuchElementException:
        Fuel.append('None')
        
        
            
    # Extracting KM Driven  
    try:
        km_driven = driver.find_element(By.XPATH,'//*[@id="vdpCard-carOverview"]/ul/li[5]/div/div/span')      
        KMs_Driven.append(km_driven.text)
    except NoSuchElementException:
        KMs_Driven.append('None')
        
        
    # Extracting Engine_Displacement  
    try:
        engine_displacement = driver.find_element(By.XPATH,'//*[@id="vdpCard-carOverview"]/ul/li[8]/div/div/span')      
        Engine_Displacement.append(engine_displacement.text)
    except NoSuchElementException:
        Engine_Displacement.append('None')
        
        
    # Extracting No_Of_Owner
    try:
        no_Of_owner = driver.find_element(By.XPATH,'//*[@id="vdpCard-carOverview"]/ul/li[7]/div/div/span')      
        No_Of_Owner.append(no_Of_owner.text)
    except  NoSuchElementException:
        No_Of_Owner.append('None')
        
        
    # Extracting Transmission
    try:
        transmission = driver.find_element(By.XPATH,'//*[@id="vdpCard-carOverview"]/ul/li[9]/div/div/span')      
        Transmission.append(transmission.text)
    except  NoSuchElementException:
        Transmission.append('None')
        
        
    # Extracting Mileage
    try:
        mileage = driver.find_element(By.XPATH,'//*[@id="vdpCard-carFeatures"]/ul/li[1]/div/div/span')      
        Mileage.append(mileage.text)
    except  NoSuchElementException:
        Mileage.append("None")
        

        
        
        
 # Extracting Torque
    try:
        torque = driver.find_element(By.XPATH,'//*[@id="vdpCard-carFeatures"]/ul/li[4]/div/div/span')      
        Torque.append(torque.text)
    except  NoSuchElementException:  
        Torque.append('None')
        
        
 # Extracting Seats
    try:
        seats = driver.find_element(By.XPATH,'//*[@id="vdpCard-carFeatures"]/ul/li[5]/div/div/span')      
        Seats.append(seats.text)
    except NoSuchElementException: 
        Seats.append('-')
```

```python
for link in Car_urls_amd:
    driver.get(link)

# Extracting Gear_Box
    try:
        gear_Box = driver.find_element(By.XPATH,'//*[@id="vdpCard-carFeatures"]/div[2]/div[3]/div/div/ul/li[1]/div/div/span')      
        Gear_Box.append(gear_Box.text)
    except NoSuchElementException:
        Gear_Box.append('-') 
        
        
        
            # Extracting Drive_Type
    try:
        drive_Type = driver.find_element(By.XPATH,'//*[@id="vdpCard-carFeatures"]/div[2]/div[3]/div/div/ul/li[2]/div/div/span')      
        Drive_Type.append(drive_Type.text)
    except NoSuchElementException:
        Drive_Type.append('-')
        
        
        
                
    # Extracting Max_Power
    try:
        max_power = driver.find_element(By.XPATH,'//*[@id="specification-wdght"]/div/div/div/ul/li[3]/div/div[2]')      
        Max_Power.append(max_power.text)
    except  NoSuchElementException:   
        Max_Power.append('None')
        
        
        
        
    # Extracting Steering_Type
    try:
        steering_Type = driver.find_element(By.XPATH,'//*[@id="vdpCard-carFeatures"]/div[2]/div[3]/div/div/ul/li[4]/div/div/span')      
        Steering_Type.append(steering_Type.text)
    except NoSuchElementException:
        Steering_Type.append('-')
        
        
    # Extracting Front_Brake_Type
    try:
        front_Brake_Type = driver.find_element(By.XPATH,'//*[@id="vdpCard-carFeatures"]/div[2]/div[3]/div/div/ul/li[6]/div/div/span')      
        Front_Brake_Type.append(front_Brake_Type.text)
    except NoSuchElementException:
        Front_Brake_Type.append('-')
        
        
        
        # Extracting Color
    try:
        color = driver.find_element(By.XPATH,'//*[@id="specification-wdght"]/div/div/div/ul/li[6]/div/div[2]')      
        Color.append(color.text)
    except NoSuchElementException:
        Color.append('-')
        
            
    # Extracting Rear_Brake_Type
    try:
        rear_Brake_Type = driver.find_element(By.XPATH,'//*[@id="vdpCard-carFeatures"]/div[2]/div[3]/div/div/ul/li[7]/div/div/span')      
        Rear_Brake_Type.append(rear_Brake_Type.text)
    except NoSuchElementException:
        Rear_Brake_Type.append('-')
        
        
    # Extracting Acceleration
    try:
        acceleration = driver.find_element(By.XPATH,'//*[@id="vdpCard-carFeatures"]/div[2]/div[3]/div/div/ul/li[9]/div/div/span')      
        Acceleration.append(acceleration.text)
    except NoSuchElementException:
        Acceleration.append('-')
        
        
    # Extracting Engine Type
    try:
        engine_Type = driver.find_element(By.XPATH,'//*[@id="vdpCard-carFeatures"]/div[2]/div[1]/div/div/ul/li[2]/div/div/span')      
        Engine_Type.append(engine_Type.text)
    except NoSuchElementException:
        Engine_Type.append('-')
        
        
    #Extracting Wheel Base
    try:
        wheel_base = driver.find_element(By.XPATH,'//*[@id="vdpCard-carFeatures"]/div[2]/div[2]/div/div/ul/li[4]/div/div/span')      
        Wheel_Base.append(wheel_base.text)
    except NoSuchElementException:
        Wheel_Base.append('-')
```

```python
# Making dataFrame for Ahmdabad cars
Delhi=pd.DataFrame({})
Delhi ['Brand']= Brand [:100]

Delhi ['Make_Year']= Make_Year [:100]
Delhi ['Fuel']=  Fuel   [:100]
Delhi ['KMs_Driven']= KMs_Driven [:100]
Delhi ['Engine_Displacement']= Engine_Displacement [:100]
Delhi ['No_Of_Owner']= No_Of_Owner [:100]
Delhi ['Transmission']= Transmission [:100]
Delhi ['Mileage']= Mileage [:100]
Delhi ['Max_Power']= Max_Power [:100]
Delhi ['Torque']= Torque [:100]

Delhi ['Seats']= Seats[:100]
Delhi ['Color']= Color [:100]
Delhi ['Gear_Box']= Gear_Box [:100]
Delhi ['Drive_Type']= Drive_Type [:100]
Delhi ['Steering_Type']= Steering_Type [:100]
Delhi ['Front_Brake_Type']= Front_Brake_Type [:100]
Delhi ['Rear_Brake_Type']= Rear_Brake_Type [:100]

Delhi ['Acceleration']= Acceleration [:100]
Delhi ['Engine_Type']= Engine_Type[:100]
Delhi ['Price']= Price [:100]
```
