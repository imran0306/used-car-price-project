```python
# scrape all Cars urls
Car_urls_amd = []
url_amd=driver.find_elements(By.XPATH,"//*/div/div[2]/div/div/a")
for car in url_amd:
        Car_urls_amd.append(car.get_attribute("href"))
```

```python
Links_amd=Car_urls_amd
len(Car_urls_amd)
```

```python
# Make Empty List of the required data
Brand= []
Model= []
Make_Year= []
Fuel=[]
KMs_Driven=[]
Engine_Displacement=[]
No_Of_Owner=[]
Transmission=[]
Mileage=[]
Max_Power=[]
Torque=[]

Seats=[]
Color=[]
Gear_Box=[]
Drive_Type=[]
Steering_Type=[]
Front_Brake_Type=[]
Rear_Brake_Type=[]

Acceleration=[]
Engine_Type=[] 
Wheel_Base = []
Price= []
```

```python
for link in Car_urls_amd:
    driver.get(link)
  
    

    
    # Extracting Price 
    try:
        price = driver.find_element(By.XPATH,'//*[@id="rf01"]/div[1]/div/div/main/div/div[1]/div[2]/div[1]/div/div/div[2]/div[1]/div[2]/div[1]/div[1]/span[2]')      
        Price.append(price.text)
   
    except NoSuchElementException:
        Price.append('None')
        
    # Extracting Brand from xpath
    try:
        brand = driver.find_element(By.XPATH,'//*[@id="rf01"]/div[1]/div/div/main/div/div[1]/div[2]/div[1]/div/div/div[2]/div[1]/div[1]/h1')      
        Brand.append(brand.text)
    except NoSuchElementException:
        Price.append('None')
        
    # Extracting make year
    try:
        year = driver.find_element(By.XPATH,'//*[@id="Overview"]/div/div/div/ul/li[1]/div/div[1]')      
        Make_Year.append(year.text)
    except NoSuchElementException:
        Make_Year.append('-')
        
        
# Extracting Fuel Type 
    try:
        fuel = driver.find_element(By.XPATH,'//*[@id="widget-Overview"]/div/div/div/div[1]/ul/li[5]/div/div')      
        Fuel.append(fuel.text)
    except NoSuchElementException:
        Fuel.append('None')
        
        
            
    # Extracting KM Driven  
    try:
        km_driven = driver.find_element(By.XPATH,'//*[@id="Overview"]/div/div/div/ul/li[4]/div/div[2]')      
        KMs_Driven.append(km_driven.text)
    except NoSuchElementException:
        KMs_Driven.append('None')
        
        
    # Extracting Engine_Displacement  
    try:
        engine_displacement = driver.find_element(By.XPATH,'//*[@id="Overview"]/div/div/div/ul/li[5]/div/div[2]')      
        Engine_Displacement.append(engine_displacement.text)
    except NoSuchElementException:
        Engine_Displacement.append('None')
        
        
    # Extracting No_Of_Owner
    try:
        no_Of_owner = driver.find_element(By.XPATH,'//*[@id="Overview"]/div/div/div/ul/li[6]/div/div[2]')      
        No_Of_Owner.append(no_Of_owner.text)
    except  NoSuchElementException:
        No_Of_Owner.append('None')
        
        
    # Extracting Transmission
    try:
        transmission = driver.find_element(By.XPATH,'//*[@id="Overview"]/div/div/div/ul/li[8]/div/div[2]')      
        Transmission.append(transmission.text)
    except  NoSuchElementException:
        Transmission.append('None')
        
        
    # Extracting Mileage
    try:
        mileage = driver.find_element(By.XPATH,'//*[@id="specification-wdght"]/div/div/div/ul/li[1]/div/div[2]')      
        Mileage.append(mileage.text)
    except  NoSuchElementException:
        Mileage.append("None")
        
        
    # Extracting Max_Power
    try:
        max_power = driver.find_element(By.XPATH,'//*[@id="specification-wdght"]/div/div/div/ul/li[3]/div/div[2]')      
        Max_Power.append(max_power.text)
    except  NoSuchElementException:   
        Max_Power.append('None')
        
        
        
 # Extracting Torque
    try:
        torque = driver.find_element(By.XPATH,'//*[@id="specification-wdght"]/div/div/div/ul/li[4]/div/div[2]')      
        Torque.append(torque.text)
    except  NoSuchElementException:  
        Torque.append('None')
        
        
            # Extracting Seats
    try:
        seats = driver.find_element(By.XPATH,'//*[@id="specification-wdght"]/div/div/div/ul/li[5]/div/div[2]')      
        Seats.append(seats.text)
    except NoSuchElementException:
        Seats.append('-')
        
        
            
    # Extracting Color
    try:
        color = driver.find_element(By.XPATH,'//*[@id="specification-wdght"]/div/div/div/ul/li[6]/div/div[2]')      
        Color.append(color.text)
    except NoSuchElementException:
        Color.append('-')
```
